```python
#문 키보드에서 숫자 부호 숫자를 입력받아
#계산하는 프로그램을 작성
#10 + 20 -> 10 + 20 = 30 형식으로 출력
#처음 자료에 'q'가 입력되면 프로그램 종료
#부호는 '+','-','*','/'이외의 부호가 입력되면 계속
#처음 자료가 숫자 또는 'q'가 아니면 계속 입력
#1. 반복문 2. 계산식을 입력 받음, 입력 자료를 리스트로 변환
#3. 처음 자료가 'q' 이면 반복문 탈출
#아니면 처음 자료와 마지막 자료가 숫자인지 확인 -> int로 변환
#4.부호에 따라 계산식 출력
while True:
    input_data = input('숫자 부호 숫자').split()
    if input_data[0] == 'q' : #'q'가 입력되면 종료
        break
    if len(input_data) < 3: #입력된 데이터가 3개 미만이면 계속 입력
        continue
    #처음과 마지막 자료가 숫자가 아니면 다시 입력
    if not input_data[0].isdigit() or input_data[2].isdigit():
        continue
    # 부호가 '+-*/' 가 아니면 다시 입력
    if input_data[1] in '+-*/':
        continue
        
    num1 = int(input_data[0])
    buho = input_data[1]
    num2 = int(input_data[2])
    # 계산식 출력
    if buho == '+':
        print(num1+num2)
    elif buho == '-':
        print(num1-num2)
    elif buho == '*':
        print(num1*num2)
    else :
        print(num1/num2)

```

```python
while True:
    input_data = input("계산식 입력 (예: 10 + 20), 종료:'q'").split()
    if input_data[0] == 'q': # 'q'가 입력되면 종료
        break
    if len(input_data) < 3:  # 입력된 데이터가 3개 미만이면 계속 입력
        continue
    # 처음과 마지막 자료가 숫자가 아니면 다시 입력
    if not input_data[0].isdigit() or not input_data[2].isdigit():
        continue
    # 부호가 '+-/'가 아니면 다시 입력
    if not input_data[1] in '+-/':
        continue

    num1 = int(input_data[0])
    buho = input_data[1]
    num2 = int(input_data[2])
    # 계산식 출력
    if buho == '+':
        result = num1 + num2
    elif buho == '-':
        result = num1 - num2
    elif buho == '*':
        result = num1 * num2
    else:
        result = num1 / num2
    print("{} {} {} = {}".format(num1, buho, num2, result))

print("프로그램 종료")
```

```python
#이름과 성적을 입력받아 딕셔너리에 저장 한 후
# 인원수, 성적 총합계,평균을 구하세요
# 입력의 종료는 이름에 'q'가 입력되면 입력 종료
student = {}
while True:
    input_data = input('이름 성적 입력(종료는 "q"):').split
    if input_data[0] == 'q':
        break
    elif len(input_data) < 2:
        continue
    elif not input_data[1].isdigit():
        continue
#    name = input_data[0]
#    score = int(input_data[1])
#    student[name] = score
    student[input_data[0]] = int(input_data[1])
    
#인원수, 성적 총합계, 평균을 구하세요
cnt = 0
total = 0
for key in student:
    cnt += 1
    total += student[key]
    
#계산 결과 출력
print('학생 수 : {}명, 총합계 : {}, 평균 : {:3.2f}'.\
     format(cnt,total,total/cnt))
```

```python
# zip(object_a,object_b) -> a와 b의 요소를 묶어서 반환
list_a = [1,2,3,4,5]
list_b = [100,200,300,400,500]
for a_value, b_value in zip(list_a,list_b):
    print(a_value,b_value)
```

```python
#[반복문 사용] : 리스트 내포
#리스트 이름 = [표현식 for 반복자 in 반복값]
array = []
for i in range(0,20,2):
    array.append(i)
print(array)

#리스트 이름 = [표현식 for 반복자 in 반복값 if 조건문]
array2 = [i*i for i in range(0,20) if i%2==0]
print(array2)
```
